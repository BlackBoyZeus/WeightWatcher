
# WeightWatcher - GPT


```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

### Import WeightWatcher


```python
import torch
import weightwatcher as ww
torch.__version__, ww.__version__
```

```python
import transformers
from transformers import OpenAIGPTModel,GPT2Model
```

```python
gpt_model = OpenAIGPTModel.from_pretrained('openai-gpt')
gpt_model.eval();

gpt2_model = GPT2Model.from_pretrained('gpt2')
gpt2_model.eval();
```

```python
watcher = ww.WeightWatcher(model=gpt_model)
gpt_details = watcher.analyze(fit='TPL')
```

```python
watcher = ww.WeightWatcher(model=gpt2_model)
gpt2_details = watcher.analyze(fit='TPL')
```

```python
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (7,5),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'xx-large',
         'xtick.labelsize':'xx-large',
         'ytick.labelsize':'xx-large'}
plt.rcParams.update(params)
```

```python
gpt_details.alpha.plot.hist(bins=100, color='red', alpha=0.5, density=True, label='gpt')
plt.legend()
plt.xlabel(r"alpha $(\alpha)$ TPL exponent")
```

```python
gpt_details.alpha.plot.hist(bins=100, color='red', alpha=0.5, density=True, label='gpt')
gpt2_details.alpha.plot.hist(bins=100, color='green', density=True, label='gpt2')
plt.legend()
plt.xlabel(r"alpha $(\alpha)$ TPL exponent")
```

```python
gpt_details.Lambda.plot.hist(bins=100, color='red', alpha=0.5, density=True, label='gpt')
gpt2_details.Lambda.plot.hist(bins=100, color='green', density=True, label='gpt2')
plt.legend()
plt.xlabel(r"Lambda $(\Lambda)$ TPL exponent")
plt.xlim([0,0.1])
plt.ylim([0,100])
```
